```python
import pickle
import pandas as pd
```

```python
with open('ensemble_model.pkl', 'rb') as file:
    model = pickle.load(file)
with open("Gradient Boosting.pkl", 'rb') as file:
    gb_model = pickle.load(file)
with open("Linear Support Vector Machines.pkl", 'rb') as file:
    lsvm_model = pickle.load(file)
with open("AdaBoost.pkl", 'rb') as file:
    ada_model = pickle.load(file)
```

```python
df = pd.read_excel('test_proteins_19k_final.xlsx')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
modified_df=df.drop(columns=['Unnamed: 0','uniprod_id','Entry Name'])
```

```python
modified_df.describe()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
column_stats = modified_df.describe().loc[['mean', 'std']]

std_threshold = 1.0  #threshold for standardization
minmax_threshold = 0.05  #threshold for normalization

columns_to_standardize = []
columns_to_normalize = []

for column in modified_df.columns:
    mean = column_stats[column]['mean']
    std = column_stats[column]['std']

    if std > std_threshold:
        columns_to_standardize.append(column)
    if np.abs(mean) > minmax_threshold:
        columns_to_normalize.append(column)

print("Columns to standardize:", columns_to_standardize)
print("Columns to normalize:", columns_to_normalize)
```

```python
# Normalize and standardize the selected columns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
for column in columns_to_normalize:
    scaler = MinMaxScaler()
    modified_df[column] = scaler.fit_transform(modified_df[[column]])

for column in columns_to_standardize:
    scaler = StandardScaler()
    modified_df[column] = scaler.fit_transform(modified_df[[column]])

```

```python
modified_df.describe()
```

```python
model.predict(modified_df)
```

```python
df['ensemble_predictions'] = model.predict(modified_df)
df['gb_predictions'] = gb_model.predict(modified_df)
df['ada_predictions'] = ada_model.predict(modified_df)
df['lsvm_predictions'] = lsvm_model.predict(modified_df)
```

```python
def final_prediction(row):
    if (row['gb_predictions'] + row['ada_predictions'] + row['lsvm_predictions']) >= 2:
        return 1
    else:
        return 0
    
df['voting_prediction'] = df.apply(final_prediction, axis=1)
```
